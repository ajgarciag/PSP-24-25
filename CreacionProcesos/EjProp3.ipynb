```python
import subprocess, psutil
```

```python
subprocess.Popen("notepad.exe") #ejecuto este proceso para eliminarlo después
```

```python
seguir=True
while seguir:
    print("\nIntroduce...\n 1 Para listar los procesos \n 2 Para eliminar un proceso \n 3 Para salir \n")
    aux = int(input())
    if aux == 1:
        for proc in psutil.process_iter():  
            try:
                processName = proc.name()
                processID = proc.pid
                processEstado = proc.status() #estado del proceso
                processPrior = proc.nice()  # obtener el valor 'nice'
                print(f"Nombre: {processName}, PID: {processID}, Estado: {processEstado}, Prioridad: {processPrior}")
            except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            # Estos errores pueden ocurrir si no se puede acceder al proceso
                processPrior = 'Acceso Denegado'
                print(f"Nombre: {processName}, PID: {processID}, Estado: {processEstado}, Prioridad: {processPrior}")
        
    elif aux == 2:
        try: 
            elimina=int(input("\n Introduce el PID del proceso que deseas eliminar: "))
            for proc in psutil.process_iter():  #Obtener el nombre y el PID de cada proceso
                processName = proc.name()
                processID = proc.pid
                if processID==elimina:
                    print(f"Vamos a eliminar el proceso con nombre {processName} y PID {processID}")
                    try:
                        proc.terminate()
                        print("Proceso terminado con éxito")
                    except:
                        print("Error al eliminar")
        except:
            print("Error al introducir el pid")
    else: 
        seguir=False
        
```
