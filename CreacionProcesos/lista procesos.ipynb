### Ejemplo para eliminar un proceso seg√∫n su PID

```python
try:
    for proc in psutil.process_iter():  #Obtener el nombre y el PID de cada proceso
        processName = proc.name()
        processID = proc.pid
        print(processName , ' ::: ', processID)
        if processID==18064:
            print(f"Vamos a eliminar el proceso con nombre {processName} y PID {processID}")
            try:
                proc.terminate()
                print("Proceso terminado con √©xito")
            except:
                print("Error al eliminar")
            
#except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess) as e:
except:
    print("error")
#procesos
```

#### Otras maneras de hacerlo y obtener m√°s informaci√≥n

```python
import psutil

try:
    for proc in psutil.process_iter(['pid', 'name', 'cpu_percent', 'memory_info', 'status', 'create_time']):  
        processID = proc.info['pid']
        processName = proc.info['name']
        cpuUsage = proc.info['cpu_percent']  # Uso de CPU
        memoryUsage = proc.info['memory_info'].rss  # Uso de memoria en bytes (RSS)
        status = proc.info['status']  # Estado del proceso (running, sleeping, etc.)
        createTime = proc.info['create_time']  # Tiempo de creaci√≥n del proceso (en timestamp)
        
        print(f"Nombre: {processName}, PID: {processID}, CPU: {cpuUsage}%, Memoria: {memoryUsage} bytes, Estado: {status}, Inicio: {createTime}")
except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess) as e:
    print(f"Error al obtener informaci√≥n del proceso: {e}")
```
