```python
import threading, random

terminar = False

def genera():
    global lista, terminar
    while not terminar:
        with cond:
            cond.wait()
            lista.append(random.randint(1,100))
            cond.notify()

def recorre():
    global lista, terminar
    while not terminar:
        with cond:
            cond.wait()
            for i in range(len(lista)):
                if lista[i]%10==0:
                    print(lista)
                    print(f"Se va a cambiar el elemento {lista[i]} por -1")
                    lista[i]=-1
                    print(lista)
            cond.notify()

def sumar():
    global lista, terminar, suma
    while not terminar:
        with cond:
            cond.wait()
            suma = sum(lista)
            if suma >= 20000:
                terminar = True
            cond.notify()
    
    return suma


lista = []
hilos = []
metodos = [genera,recorre,sumar]
cond = threading.Condition()

for x in metodos:
    t = threading.Thread(target=x)
    hilos.append(t)
    t.start() 
with cond: 
    cond.notify()
for hilo in hilos:
    hilo.join()


print(f"La lista final es {lista}\ny la suma es {suma}")
```
